```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Dicionário de faturamento
dict_revenue = {
    'date_ref': [
        '2023-01-01',
        '2020-02-01',
        '2021-03-01',
        '2022-04-01',
        '2023-05-01',
        '2023-06-01',
        '2020-07-01',
        '2021-08-01',
        '2022-09-01',
        '2023-10-01',
        '2022-11-01',
        '2023-12-01',
    ],
    'value': [
        400000,
        890000,
        760000,
        430000,
        920000,
        340000,
        800000,
        500000,
        200000,
        900000,
        570000,
        995000,
    ]
}
```

```python
# criar Dataframe
df_revenue = pd.DataFrame.from_dict(dict_revenue)
```

```python
# Converter a coluna 'date_ref' para o tipo datetime
df_revenue['date_ref'] = pd.to_datetime(df_revenue['date_ref'])

# Ordenar o DataFrame pela coluna 'date_ref'
df_revenue.sort_values(by='date_ref', inplace=True)

df_revenue
```

```python
# exibir info do Dataframe
df_revenue.info()
```

```python
# gerar média de faturamento
revenue_mean = df_revenue["value"].mean()
print(f'Média de Faturamento: ${revenue_mean:.2f}')
```

```python
# criar um gráfico de barras vertical mostrando o mês de referência e o valor
plt.figure(figsize=(12, 6))
bars = plt.bar(df_revenue['date_ref'],
               df_revenue['value'], color='skyblue', width=20)
plt.title('Faturamento por Mês', fontsize=14)
plt.xlabel('Mês de Referência', fontsize=12)
plt.ylabel('Faturamento', fontsize=12)
plt.xticks(rotation=45, fontsize=9)
plt.yticks(fontsize=9)
plt.ticklabel_format(axis='y', style='plain')
plt.gca().yaxis.grid(True)  # Adiciona grade no eixo y

# Adiciona os valores de cada barra no topo
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height,
             height, ha='center', va='bottom', fontsize=7)

plt.tight_layout()
plt.show()
```

```python
# Criar um gráfico de linhas mostrando o mês de referência e o valor.
plt.figure(figsize=(12, 6))
plt.plot(df_revenue['date_ref'], df_revenue['value'],
         marker='o', color='green', linestyle='-')
plt.title('Faturamento por Mês', fontsize=14)
plt.xlabel('Mês de Referência', fontsize=12)
plt.ylabel('Faturamento', fontsize=12)
plt.xticks(rotation=45, fontsize=9)
plt.yticks(fontsize=9)
plt.ticklabel_format(axis='y', style='plain')
plt.gca().yaxis.grid(True)  # Adiciona grade no eixo y

# Adiciona os valores de cada ponto no gráfico
for x, y in zip(df_revenue['date_ref'], df_revenue['value']):
    plt.text(x, y, f'{y}', ha='right', va='bottom', fontsize=7)

plt.tight_layout()
plt.show()
```
